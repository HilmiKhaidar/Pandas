```python
import pandas as pd

#NO 1 
data = pd.read_csv('data.csv')  
data

```

```python
# NO 2
# membuat dataFrame
df = pd.DataFrame(data)

# menghitung total produksi sampah untuk tahun tertentu
tahun_tertentu = 2018  # disini kita bebas untuk memasukan tahun yang kita / saya inginkan
total_sampah = 0

# terus ini untuk yang menggunakan itterrows
for index, row in df.iterrows():
    if row['tahun'] == tahun_tertentu:
        total_sampah += row['jumlah_produksi_sampah']

# untuk menghasilkan outputnya
print(f"Total produksi sampah di Jawa Barat pada tahun {tahun_tertentu}: {total_sampah} ton")
```

```python
# jumlah data pertahun menggunakan itterrows
total_per_tahun = {}
for index, row in df.iterrows():
    tahun = row['tahun']
    jumlah_sampah = row['jumlah_produksi_sampah']
    
    if tahun not in total_per_tahun:
        total_per_tahun[tahun] = 0
    total_per_tahun[tahun] += jumlah_sampah

# untuk menghasilkan output
for tahun, total_sampah in total_per_tahun.items():
    print(f"Tahun {tahun}: {total_sampah} ton")
```

```python
# jadi di bagian ini untuk menginisialisasi dictionary untuk menyimpan total per kota / kabupaten dan tahun
total_per_kota_tahun = {}

# iterasi menggunakan iterrows
for index, row in df.iterrows():
    kota = row['nama_kabupaten_kota']
    tahun = row['tahun']
    jumlah_sampah = row['jumlah_produksi_sampah']
    
    if (kota, tahun) not in total_per_kota_tahun:
        total_per_kota_tahun[(kota, tahun)] = 0
    total_per_kota_tahun[(kota, tahun)] += jumlah_sampah

# ini buat output
for (kota, tahun), total_sampah in total_per_kota_tahun.items():
    print(f"Kota/Kabupaten: {kota}, Tahun: {tahun}, Total Sampah: {total_sampah} ton")
```

```python
result_df = pd.DataFrame(list(total_per_kota_tahun.items()), columns=['Kota/Tahun', 'Total Sampah'])

#  buat ekspor ke CSV 
result_df.to_csv('hasil.csv', index=False)

# ini untuk ekspor ke Excel 
result_df.to_excel('hasil.xlsx', index=False)
```
